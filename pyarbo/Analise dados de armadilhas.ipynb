```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df_arm = pd.read_csv('https://raw.githubusercontent.com/eduardocorrearaujo/arbo-fronteiras/main/data/mosq_aaeg_trap-2017_2022.csv')

df_arm.set_index('date', inplace = True);

df_arm.index = pd.to_datetime(df_arm.index)

df_arm.sort_index(inplace = True)

df_arm.head(20)
```

```python
df_arm.loc[(df_arm.n_traps != 0) & (df_arm.index >= '2017-11-01') & (df_arm.index <= '2018-02-01')]
```

Agregando os dados a cada 2 meses:

```python
df_arm_bi = df_arm.resample('2M').sum()

df_arm_bi.head()
```

```python
df_arm_bi.loc[df_arm_bi.n_traps >= 2500]
```

```python
fig, ax = plt.subplots()

ax.plot(df_arm_bi.m_aaeg_f_m + df_arm_bi.m_aaeg_f_v, marker = 'o')

ax.grid()

ax.set_title('Mosquitos fmeas capturados (vivos e mortos)')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot((df_arm_bi.m_aaeg_f_m + df_arm_bi.m_aaeg_f_v)/(df_arm_bi.n_traps), marker = 'o')

ax.grid()

ax.set_title('Mosquitos fmeas capturados (vivos e mortos)')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.plot(df_arm_bi.n_traps, marker = 'o')

ax.grid()

ax.set_title('Nmero de armadilhas')

plt.show()
```
