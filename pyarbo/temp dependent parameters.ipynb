```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def briere(T, a, b, c):
    
    return a*T*(T-b)*((c-T)**(1/2))

def quad(T, a, b, c):
   
    return a*T*(T-b)*(T-c)

def b(a, T):
    
    values = []
    for temp in T:
        if (temp > 34) | (temp < 18):
            v = 0
        else: 
            v = briere(temp, a = a, b=18, c = 34)
            
        values.append(v)
    
    return values


def beta(a, T):
    
    values = []
    for temp in T:
        if (temp > 34) | (temp < 18):
            v = 0
        else: 
            v = briere(temp, a = a, b=18, c = 34)
            
        values.append(v)
    
    return values
```

$$b(T) = aT(T-T_{min})[(T_{máx}-T)^{1/2}]$$

Seguindo Mordecai, $$T_{min} = 18°C, T_{max} = 34°C.$$

```python
T = np.arange(-1.7, 41.7, 0.1)

fig, ax = plt.subplots()

ax.plot(T, b(a =0.00121, T = T))

ax.grid()

ax.set_title('Bitting rate / Taxa de contágio')

ax.set_xlabel('Temperatura °C')

ax.set

plt.show()
```

Variação da função de briere:

```python
T = np.arange(18, 34.1, 0.1)

fig, ax = plt.subplots()

for a in np.arange(0.0001, 0.0016, 0.0002):

    ax.plot(T, briere(T, a = a, b = 18, c = 34), label = f'a = {a}')

ax.grid()

ax.set_title('Bitting rate')

ax.legend()

plt.show()
```
